```python
import pandas as pd
import os
import matplotlib.pyplot as plt
print(os.listdir())
list_value_PPO =[]
list_value_ACKTR =[]
list_value_DDPG =[]
list_value_TD3 =[]
list_value_Ensemble =[]
plt.rcParams["figure.figsize"] = (20,12)

%cd /content/Forex-Trading-Automation-with-Deep-Reinforcement-Learning/results/PPO
list_file = []
for file in os.listdir():
  cl = []
  if file[-3:] == 'csv' and file[8:13] == 'value':
    list_file.append(file)
list_file = sorted(list_file)
print(list_file)
for file in list_file:
    f = pd.read_csv(f'{file}')
    list_att = f['0'].tolist()
    for attribute in list_att:
      list_value_PPO.append(attribute)
count = len(list_value_PPO)
for i in range(0, count, 1):
  cl.append(i)
print(cl)
%cd /content/Forex-Trading-Automation-with-Deep-Reinforcement-Learning/results/TD3
list_file = []
for file in os.listdir():
  cl = []
  if file[-3:] == 'csv' and file[8:13] == 'value':
    list_file.append(file)
list_file = sorted(list_file)
print(list_file)
for file in list_file:
    f = pd.read_csv(f'{file}')
    list_att = f['0'].tolist()
    for attribute in list_att:
      list_value_TD3.append(attribute)
count = len(list_value_TD3)
for i in range(0, count, 1):
  cl.append(i)
%cd /content/Forex-Trading-Automation-with-Deep-Reinforcement-Learning/results/DDPG
list_file = []
for file in os.listdir():
  cl = []
  if file[-3:] == 'csv' and file[8:13] == 'value':
    list_file.append(file)
list_file = sorted(list_file)
print(list_file)
for file in list_file:
    f = pd.read_csv(f'{file}')
    list_att = f['0'].tolist()
    for attribute in list_att:
      list_value_DDPG.append(attribute)
count = len(list_value_DDPG)
for i in range(0, count, 1):
  cl.append(i)
%cd /content/Forex-Trading-Automation-with-Deep-Reinforcement-Learning/results/ACKTR
list_file = []
for file in os.listdir():
  cl = []
  if file[-3:] == 'csv' and file[8:13] == 'value':
    list_file.append(file)
list_file = sorted(list_file)
print(list_file)
for file in list_file:
    f = pd.read_csv(f'{file}')
    list_att = f['0'].tolist()
    for attribute in list_att:
      list_value_ACKTR.append(attribute)
count = len(list_value_ACKTR)
for i in range(0, count, 1):
  cl.append(i)
%cd /content/Forex-Trading-Automation-with-Deep-Reinforcement-Learning/results/ensemble
list_file = []
for file in os.listdir():
  cl = []
  if file[-3:] == 'csv' and file[8:13] == 'value':
    list_file.append(file)
list_file = sorted(list_file)
print(list_file)
for file in list_file:
    f = pd.read_csv(f'{file}')
    list_att = f['0'].tolist()
    for attribute in list_att:
      list_value_Ensemble.append(attribute)
count = len(list_value_Ensemble)
for i in range(0, count, 1):
  cl.append(i)
plt.plot(cl, list_value_PPO, label = "PPO")
plt.plot(cl, list_value_TD3, label = "TD3")
plt.plot(cl, list_value_DDPG, label = "DDPG")
plt.plot(cl, list_value_ACKTR, label = "ACKTR")
plt.plot(cl, list_value_Ensemble, label = "Ensemble method", marker ='o', markersize = 3)
plt.title('Số dư tài khoản của các \'Trader\' Deep Reiforcement Learning sau trong thời điểm 07/2020 tới 07/2022')
plt.xlabel('Khoảng thời gian (Ngày)')
plt.ylabel('Số dư tài khoản (1000000 USD)')
plt.legend()
plt.show()
df = pd.DataFrame({'PPO':list_value_PPO,
                   'TD3':list_value_TD3,
                   'DDPG':list_value_DDPG,
                   'ACKTR':list_value_ACKTR,
                   'Ensemble': list_value_Ensemble})
df
```
